Steps:
Load the dataset into a Pandas DataFrame and explore the data.

Create a line plot showing the monthly revenue over the two-year period. Customize the plot to make it publication-quality.

Create a bar plot showing the total number of orders per month. Customize the plot to make it publication-quality.

Create a scatter plot showing the relationship between the number of orders and the average order value. Customize the plot to make it publication-quality.

Create a subplot containing all three visualizations created in steps 2-4. Adjust the layout and size of the subplot to make it visually appealing.

Save the subplot as a high-resolution image file.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("jumta.csv")
data.head()
```

```python
y= data['Revenue']
x = data['Month']

plt.title("Monthly Revenue (2020-2021)")
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.xticks(rotation='vertical')
fig = plt.gcf()
fig.set_size_inches(12, 6)
plt.grid(color='gray', linestyle='-', linewidth=0.5)
plt.plot(x,y, marker='o', label='Revenue')
plt.legend(loc="upper left")
```

```python
y= data['Orders']
x = data['Month']

plt.title("Monthly Orders (2020-2021)")
plt.xlabel('Month')
plt.ylabel('Number of Orders')
plt.xticks(rotation='vertical')
fig = plt.gcf()
fig.set_size_inches(12, 6)
plt.bar(x,y, label='Orders', color="orange")
plt.legend(loc="upper left")
```

```python
y= data['Average_Order_Value']
x = data['Orders']

plt.title("Number of Orders vs Average Order Value (2020-2021)")
plt.xlabel('Number of Orders')
plt.ylabel('Average Order Value')
fig = plt.gcf()
fig.set_size_inches(12, 6)
plt.grid(color='gray', linestyle='-', linewidth=0.5)
plt.scatter(x,y, label='Order Value', color="green")
plt.legend(loc="upper right")
```

```python
x = data['Month']
x1 = data['Orders']
y = data['Revenue']
y1 = data['Orders']
y3 = data['Average_Order_Value']

fig, (ax1, ax2, ax3) = plt.subplots(nrows=3, ncols=1, figsize=(22, 16), sharex=False)


ax1.plot(x,y, marker='o', label='Revenue')

ax1.set_title("Monthly Revenue (2020-2021)")
ax1.set_ylabel('Revenue')
ax1.set_xlabel('Month')
ax1.grid(color='gray', linestyle='-', linewidth=0.5)
ax1.legend(loc="upper left")

ax2.bar(x,y1, label='Orders', color="orange")

ax2.set_title("Monthly Revenue (2020-2021)")
ax2.set_ylabel('Number of Orders')
ax2.set_xlabel('Month')

ax2.legend(loc="upper left")

ax3.scatter(x1,y3, label='Order Value', color="green")

ax3.set_title("Number of Orders vs Average Order Value (2020-2021)")
ax3.set_ylabel('Average Order Value')
ax3.set_xlabel('Number of Orders')
ax3.grid(color='gray', linestyle='-', linewidth=0.5)
ax3.legend(loc="upper right")

plt.tight_layout()
plt.show()

plt.savefig('assignment.png', dpi=300)
```
